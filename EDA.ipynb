```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # type: ignore

df= pd.read_csv(r"C:\Users\sarvesh\OneDrive\文档\Desktop\sample_churn_data.csv")
df.head()
df.shape
df.info()
df.describe()
df.isnull().sum()
print(df.columns)

sns.countplot(x='Churn', data= df)
plt.title('Churn count')
plt.xticks([0,1], ['no', 'yes'])
plt.show()

sns.histplot(df['MonthlyCharges'], kde=True)
plt.title('Distribution of Monthly charges')
plt.show()

sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title('Monthly Charges by Churn')
plt.xticks([0,1], ['no', 'yes'])
plt.show()

sns.histplot(data=df, x='Tenure', hue= 'Churn', multiple='stack')
plt.title('Tenure vs Churn')
plt.show()

contract_churn = df.groupby('HasContract')['Churn'].mean().reset_index()
sns.barplot(x='HasContract', y='Churn', data=contract_churn)
plt.title('Churn Rate by Contract')
plt.xticks([0, 1], ['No Contract', 'Contract'])
plt.ylabel('Churn Rate')
plt.show()

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('correlation matrix')
plt.show()
```
